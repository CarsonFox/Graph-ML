The PageRank equation is as follows.

$r = \beta P^Tr + (1-\beta)/n*1_n$

where $r$ represents the $n \times 1$ PageRank vector with each element $r_i$ representing the PageRank value of node $i$, $P$ represents the $n \times n$ transition probability matrix with each element $P_{i,j}$ representing the transition probability from node $i$ to node $j$, $P^T$ represents the transpose of $P$, $\beta$ represents decay factor, $1_n$ represents a n×1 vector of all 1’s, and $n$ represents the number of nodes in the graph.
